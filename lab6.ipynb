```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

independent_feature = X[:, 0]
dependent_feature = X[:, 1]

# Create a Linear Regression model
model = LinearRegression()

# Reshape the independent feature to 2D
independent_feature = independent_feature.reshape(-1, 1)

# Fit the model
model.fit(independent_feature, dependent_feature)

# Make predictions
predicted_values = model.predict(independent_feature)

mse = mean_squared_error(dependent_feature, predicted_values)
mse
```
